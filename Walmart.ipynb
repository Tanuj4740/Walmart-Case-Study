```python
import numpy as np
female_customers=df[df['Gender']=='F']
male_customers=df[df['Gender']=='M']
avg_spending_male = male_customers['Purchase'].mean()
avg_spending_female = female_customers['Purchase'].mean()

# Sample size
n_female = len(female_customers)
n_male = len(male_customers)

# Standard deviation of the sample
std_dev_female = female_customers['Purchase'].std()
std_dev_male = male_customers['Purchase'].std()

# Standard error of the mean
sem_female = std_dev_female / np.sqrt(n_female)
sem_male = std_dev_male / np.sqrt(n_male)
critical_value = 1.96
```

```python
# Confidence interval for female customers
ci_female_lower = avg_spending_female - critical_value * sem_female
ci_female_upper = avg_spending_female + critical_value * sem_female

# Confidence interval for male customers
ci_male_lower = avg_spending_male - critical_value * sem_male
ci_male_upper = avg_spending_male + critical_value * sem_male
```

```python
# Print the results
print("Confidence Interval for Average Spending of Female Customers:")
print("Lower Bound:", ci_female_lower)
print("Upper Bound:", ci_female_upper)
print()

print("Confidence Interval for Average Spending of Male Customers:")
print("Lower Bound:", ci_male_lower)
print("Upper Bound:", ci_male_upper)
```

```python
 import numpy as np

# Function to calculate confidence interval
def calculate_ci(data, confidence_level=0.95):
    # Calculate sample mean and standard error
    sample_mean = np.mean(data)
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))

    # Calculate the critical value based on the confidence level
    if confidence_level == 0.90:
        z_score = 1.645
    elif confidence_level == 0.95:
        z_score = 1.96
    elif confidence_level == 0.99:
        z_score = 2.576
    else:
        raise ValueError("Unsupported confidence level. Please choose from 0.90, 0.95, or 0.99.")

    # Calculate margin of error
    margin_of_error = z_score * std_error

    # Calculate confidence interval
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return ci_lower, ci_upper

# Set confidence levels
confidence_levels = [0.90, 0.95, 0.99]

# Set sample sizes
sample_sizes = [50, 100, 200, 500, 1000]

# Initialize dictionaries to store results
ci_results_female = {}
ci_results_male = {}

# Loop through sample sizes and confidence levels
for sample_size in sample_sizes:
    for confidence_level in confidence_levels:
        # Sample from the original data for female and male customers
        sample_female = np.random.choice(female_customers['Purchase'], size=sample_size, replace=True)
        sample_male = np.random.choice(male_customers['Purchase'], size=sample_size, replace=True)

        # Calculate confidence intervals using CLT
        ci_lower_female, ci_upper_female = calculate_ci(sample_female, confidence_level)
        ci_lower_male, ci_upper_male = calculate_ci(sample_male, confidence_level)

        # Store results
        ci_results_female[(sample_size, confidence_level)] = (ci_lower_female, ci_upper_female)
        ci_results_male[(sample_size, confidence_level)] = (ci_lower_male, ci_upper_male)

# Print results
print("Confidence Intervals for Average Spending of Female Customers:")
for key, value in ci_results_female.items():
    print("Sample Size:", key[0], "| Confidence Level:", key[1], "| Confidence Interval:", value)

print("\nConfidence Intervals for Average Spending of Male Customers:")
for key, value in ci_results_male.items():
    print("Sample Size:", key[0], "| Confidence Level:", key[1], "| Confidence Interval:", value)
```

```python
import matplotlib.pyplot as plt

# Plot confidence intervals
plt.figure(figsize=(10, 6))

# Female customers
for key, value in ci_results_female.items():
    plt.plot([key[0], key[0]], [value[0], value[1]], color='blue', linewidth=2)

# Male customers
for key, value in ci_results_male.items():
    plt.plot([key[0], key[0]], [value[0], value[1]], color='red', linewidth=2)

plt.xlabel('Sample Size')
plt.ylabel('Average Spending')
plt.title('Confidence Intervals for Average Spending of Male and Female Customers')
plt.legend(['Female (90%)', 'Female (95%)', 'Female (99%)', 'Male (90%)', 'Male (95%)', 'Male (99%)'])
plt.grid(True)
plt.show()
```
